```
# Trips table and data type for each column

from pyspark.sql.types import *

schema = StructType([
	StructField("vendor_id", StringType()),
	StructField("pickup_datetime", TimestampType()),
	StructField("dropoff_datetime", TimestampType()),
	StructField("passenger_count", IntegerType()),
	StructField("trip_distance", DoubleType()),
	StructField("pickup_longitude", DoubleType()),
	StructField("pickup_latitude", DoubleType()),
	StructField("rate_code", DoubleType()),
	StructField("store_and_fwd_flag", StringType()),
	StructField("dropoff_longitude", DoubleType()),
	StructField("dropoff_latitude", DoubleType()),
	StructField("payment_type", StringType()),
	StructField("fare_amount", DoubleType()),
	StructField("surcharge", DoubleType()),
	StructField("tip_amount", DoubleType()),
	StructField("tolls_amount", DoubleType()),
	StructField("total_amount", DoubleType())
])

trips = spark.read.json('/data/{}/data-sample_*'.format(MOUNT_NAME), schema = schema)
trips.createOrReplaceTempView("trips")
display(trips.select("*").limit(10))
```

```
%sql

/*Average trip distance for trips with 2 passengers or less*/

SELECT format_number(avg(trip_distance), 2) as Distance 
FROM trips
WHERE passenger_count <= 2
```

```
%sql

/*Three greatest revenues from 2009 to 2012*/

SELECT first(name) as Company_Name, format_number(sum(total_amount), 2) as Revenue
FROM trips, vendors
WHERE trips.vendor_id = vendors.vendor_id
GROUP BY trips.vendor_id
ORDER BY sum(total_amount) DESC
LIMIT 3
```

```
def monthToName(n):
  return {
    1: "Jan", 2: "Fev", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun", 
    7: "Jul", 8: "Aug", 9: "Sep", 10:"Oct", 11:"Nov", 12:"Dec"
  }[n]

spark.udf.register("monthToName", monthToName, StringType())
```

```
%sql

/*Quantity of trips per month along the 4 years*/

SELECT count(*) as Trips_per_Month, monthToName(month(dropoff_datetime)) as Month
FROM trips, payments
WHERE trips.payment_type = payments.A AND payments.B LIKE "Cash"
GROUP BY month(dropoff_datetime)
ORDER BY month(dropoff_datetime) ASC
```

```
%sql

/*Sum of tips per day of Aug, Sep and Oct of 2012*/

SELECT sum(tip_amount) as Tips_Per_Day, concat_ws('-','2012', concat_ws('-', month(dropoff_datetime), day(dropoff_datetime))) as Date
FROM trips
WHERE year(dropoff_datetime) = 2012 AND month(dropoff_datetime) IN (8,9,10)
GROUP BY month(dropoff_datetime), day(dropoff_datetime)
ORDER BY month(dropoff_datetime), day(dropoff_datetime) ASC
```

```
%sql

/*Average travel time on weekends in seconds*/

SELECT format_number(avg(unix_timestamp(dropoff_datetime) - unix_timestamp(pickup_datetime)), 2) as Time
FROM trips
WHERE dayofweek(dropoff_datetime) IN (1,7)
```

```
from pyspark.sql.functions import *

# Reading dayly trips files as a stream of data and writing on a table

streamingInputDF = (
  spark
    .readStream                       
    .schema(schema)                   # Set the schema of the JSON data
    .option("maxFilesPerTrigger", 2)  # Treat a sequence of files as a stream by picking one file at a time
    .json("data/{}/2009_dayly/*/*.json".format(MOUNT_NAME))
)

streamingInputDF.createOrReplaceTempView("{}".format(MOUNT_NAME))
```

```
%sql

/* Grouping streamed data by payment type according payments table and by date to analyze the revenue */

SELECT 
  sum(total_amount) as revenue, 
  payments.B as payment_type, 
  concat_ws(', ', monthToName(month(dropoff_datetime)), dayofmonth(dropoff_datetime)) as date
FROM nyctaxitrips, payments
WHERE nyctaxitrips.payment_type = payments.A
GROUP BY payments.B, month(dropoff_datetime), dayofmonth(dropoff_datetime)
ORDER BY month(dropoff_datetime), dayofmonth(dropoff_datetime)
```
